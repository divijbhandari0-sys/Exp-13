```python
#Preprocessing on a dataset -1
import pandas as pd

data = {
    "Roll No": [101, 102, 103, 104, 105],
    "Name": ["Asha", "Rohan", "Meera", "Karan", "Anil"],
    "Age": [20, "-", 21, 20, "22"],
    "Marks": [85, 90, "-", 88, 76],
    "Department": ["CSE", "IT", "CSE", "cse", "E&TC"],
    "Admission_Date": ["01-06-2023", "03-06-2023", "05-06-2023", "07-06-2023", "10/06/23"]
}

df = pd.DataFrame(data)
df
```

```python
#Step 2 replcae '-' with nan
df.replace("-",np.nan, inplace=True)
df
```

```python
#convert coloums to numeric where needed
df["Age"] = pd.to_numeric(df["Age"], errors="coerce")
df
```

```python
##convert coloums to numeric where needed
df["Marks"] = pd.to_numeric(df["Marks"], errors="coerce")
df
```

```python
#step 3 check missing values
print("\nMissing Values:")
print(df.isna().sum())
```

```python
#fill missing age with mean
df["Age"].fillna(df["Age"].mean(), inplace=True)
df
```

```python
df["Marks"].fillna(df["Marks"].median(), inplace=True)
df
```

```python
#step 5 fix inconsistent text
df["Department"] = df["Department"].str.upper()
df
```

```python
#step6 convert admission_date to standard datetime formate
df["Admission_Date"] = pd.to_datetime(df["Admission_Date"], format="mixed",dayfirst=True,errors="coerce",)
df

```
